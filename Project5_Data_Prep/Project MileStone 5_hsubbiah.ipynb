## Reading csv file from cleaned api data file and Load to  cleaneddata_from_flat file table

```python
#Reading csv file from cleaned api data file and Load to cleaneddata_from_flat file table :
df_flatfile=pd.read_csv("cleaneddata_from_flatfile.csv")
df_flatfile

df_flatfile = df_flatfile.drop('Unnamed: 0', axis=1)
df_flatfile

```

```python
column_names = list(df_flatfile.columns.values)
column_names
```

```python
#loading to table cleaneddata_from_flatfile
conn = sqlite3.connect('final_milestone_db.sqlite')
df_flatfile.to_sql('cleaneddata_from_flatfile', conn, if_exists='replace', index=False)


query = "SELECT * FROM cleaneddata_from_flatfile"
df1 = pd.read_sql_query(query, conn)
conn.close()
df1
```

```python

#merges two tables by rows,as merging by column added lot NA values to table.
conn = sqlite3.connect('final_milestone_db.sqlite')


query = "SELECT f.magnitude,f.country,f.location,f.depth,f.mmi,f.date_time,f.year FROM cleaneddata_from_flatfile f inner JOIN cleaneddata_from_website w ON w.country=f.country UNION ALL select Magnitude,Country,Location,Depth,MMI,Date,Year from cleaneddata_from_website"
df = pd.read_sql_query(query, conn)
df.to_sql('merged_data1', conn, if_exists='replace', index=False)

query = "SELECT * FROM merged_data1"
df_mergedtable1 = pd.read_sql_query(query, conn)

df_mergedtable1


```

```python
conn = sqlite3.connect('final_milestone_db.sqlite')

query ="SELECT f.magnitude,f.date_time,f.cdi,f.mmi,f.alert,f.tsunami,f.sig,f.magType,f.depth,f.latitude,f.longitude,f.location,f.country FROM cleaneddata_from_flatfile f UNION ALL select a.mag,a.time,a.cdi,a.mmi,a.alert,a.tsunami,a.sig,a.magType,a.depth,a.lattitude,a.longitude,a.Location,a.Country from cleaneddata_from_api a"
df = pd.read_sql_query(query, conn)
df.to_sql('merged_data2', conn, if_exists='replace', index=False)

query = "SELECT * FROM merged_data2"
df_mergeddata2 = pd.read_sql_query(query, conn)

df_mergeddata2
```

# Visualization 

```python
def fig_p(data):
    series= pd.Series(data).value_counts().sort_index()
    series.plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
country_list = df1["country"].tolist()
fig_p(country_list)
plt.ylabel("Count")
plt.title("Country with most EarthQuake")
```

```python
The country with most volcanoes is: Indonesia
```

```python

plt.figure(figsize=(10,5))
fig_p(df_mergedtable1['year'])
plt.ylabel("Count")
plt.title("Number of earthquakes by year")
```
